```python
import pandas as pd
```

```python
filename = "data.json"
```

```python
df1 = pd.read_json(filename)
```

```python
df2 = df1.drop(["address", "contact", "depth", "description", "email", "url"], axis=1)
```

```python
df3 = df2.set_index("slug")
```

```python
df3.head()
```

```python
# meiste anfragen an:
most = df3.sort_values(["number_of_requests"], ascending=False)
most.head(10)
```

```python
category = []

for value in df3.categories:
    try:
        category.append(value[0]["slug"])
    except:
        category.append("keine")

df3.categories = category
```

```python
df3.head()
```

```python
classification = []

for value in df3.classification:
    try:
        classification.append(value["slug"])
    except:
        classification.append("keine")

df3.classification = classification
```

```python
df3.head()
```

```python
df3.loc["abgeordnetenhaus-von-berlin", "jurisdiction"]
```

```python
bundesland = []

for value in df3.jurisdiction:
    try:
        bundesland.append(value["slug"])
    except:
        bundesland.append("unbekannt")

df3["bundesland"] = bundesland
```

```python
bundesland_id = []

for value in df3.jurisdiction:
    try:
        bundesland_id.append(value["id"])
    except:
        bundesland_id.append("unbekannt")

df3["bundesland_id"] = bundesland_id
```

```python
df3.drop(["jurisdiction"], axis=1)
```

```python
region = []

for value in df3.jurisdiction:
    try:
        region.append(value["region"])
    except:
        region.append("unbekannt")

df3["region"] = region
df3.drop(["region"], axis=1)
```

```python
df_clean = df3.drop(["laws", "jurisdiction"], axis=1)
```
