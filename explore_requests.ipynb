```python
import pandas as pd
filename = "data_req.json"

reqs = pd.read_json(filename)
```

```python
reqs2 = reqs.drop(["checked", "is_foi", "law", "resource_uri", "summary", "title", "user"], axis=1)
```

```python
reqs2 = reqs2.set_index("slug")
```

```python
reqs2.head()
```

```python
reqs3 = reqs2.drop(["jurisdiction", "same_as_count"], axis=1)
```

```python
reqs3.head()
```

```python
# teuerste anfragen
expensive = reqs3.sort_values(["costs"], ascending=False)
expensive.head(19)
```

```python
# public body id

public_body_id = []

for value in reqs3.public_body:
    try:
        public_body_id.append(value["id"])
    except:
        public_body_id.append("keine")

reqs3["public_body_id"] = public_body_id

```

```python
# und slug dazu

public_body_slug = []

for value in reqs3.public_body:
    try:
        public_body_slug.append(value["slug"])
    except:
        public_body_slug.append("keiner")

reqs3["public_body_slug"] = public_body_slug
```
