# AVERAGE COSTS

```python
requests['costs'].describe()
```

```python
costs_per_body = requests.groupby('public_body_slug')['costs'].sum() # costs of each public body
type(costs_per_body)
```

```python
requests_per_body = requests.groupby('public_body_slug').count() # number of requests per amt, e.g. "slug" (ist immer ausgefüllt)
# achtung: stimmt nicht mit "publicbody" überein! (abgeordnetenhaus-von-berlin -> 30 vs. 27)

requests_per_body = requests_per_body["costs"]
type(requests_per_body)
```

```python
new_df = pd.concat([costs_per_body, requests_per_body], axis=1)
new_df.columns = ['total_costs', 'total_requests']
```

```python
# now create new column: average costs (total_costs/total_requests)
new_df["average_costs"] = new_df["total_costs"] / new_df["total_requests"]
new_df.head()
```

```python
new_df_final = new_df.sort_values('average_costs', ascending=False) #!!!
new_df_final
```

```python
# who is more expensive than average?
average = new_df_final.mean()
average
```

```python
# select relevant data
plot_data = new_df_final["average_costs"].head(30)
plot_data.plot(kind="bar")
plt.title("Average Costs of a Request to the Most Expensive Public Bodies")
```

# EXPENSIVE BUNDESLÄNDER

```python
pb = publicbody.set_index("slug")
pb.head()
```

```python
pb = pb["bundesland"]
pb
```

```python
w_bl = new_df
w_bl["average_costs"] = w_bl["total_costs"] / w_bl["total_requests"]
w_bl = pd.concat([w_bl, pb], axis=1, sort=True)
w_bl.head()
```

```python
w_bl.sort_values('average_costs', ascending=False).head()
```

```python
# now calculate average costs per bundesland:
bl = w_bl.groupby("bundesland").sum()
bl = bl.sort_values("total_requests", ascending=False)
bl = bl.drop("average_costs", axis=1)
bl.head()
```

```python
bl["average_costs"] = bl["total_costs"] / bl["total_requests"]
bl = bl.sort_values("average_costs", ascending=False)
bl.head(30) # !!!!!!
```

```python
import matplotlib.pyplot as plt
```

```python
# select relevant data
plot_data = bl["average_costs"].head(20)
plot_data.plot(kind="bar")
plt.title("Average Expenses per Request")
```

# Bundesland, an das die meisten Anfragen gestellt wurden:

```python
most_req = bl.sort_values("total_requests", ascending=False)
plot_data = most_req["total_requests"].head(20)

plot_data.plot(kind="bar")
plt.title("Highest Numbers of Requests to a Verwaltungseinheit")
```

# teuerstes ressort

```python
# categories, classification
neww = publicbody.set_index("slug")
new = neww["categories"]

with_cats = pd.concat([new_df_final, new], axis=1, sort=True)

with_cats.head()
```

```python
cost_of_cats = with_cats.groupby("categories").sum()
cost_of_cats = cost_of_cats.drop("average_costs", axis=1)
cost_of_cats["average_costs"] = cost_of_cats["total_costs"] / cost_of_cats["total_requests"]
cost_of_cats = cost_of_cats.sort_values("average_costs", ascending=False)

cost_of_cats.head()
```

```python
plot_cats = cost_of_cats["average_costs"].head(30)

plot_cats.plot(kind="bar")
plt.title("Highest average costs for requests to a certain category")
```
