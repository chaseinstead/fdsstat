# AVERAGE COSTS

```python
requests['costs'].describe()
```

```python
costs_per_body = requests.groupby('public_body_slug')['costs'].sum() # costs of each public body
type(costs_per_body)
```

```python
requests_per_body = requests.groupby('public_body_slug').count() # number of requests per amt, e.g. "slug" (ist immer ausgefüllt)
# achtung: stimmt nicht mit "publicbody" überein! (abgeordnetenhaus-von-berlin -> 30 vs. 27)

requests_per_body = requests_per_body["costs"]
type(requests_per_body)
```

```python
new_df = pd.concat([costs_per_body, requests_per_body], axis=1)
new_df.columns = ['total_costs', 'total_requests']
```

```python
# now create new column: average costs (total_costs/total_requests)
new_df["average_costs"] = new_df["total_costs"] / new_df["total_requests"]
new_df.head()
```

```python
new_df.sort_values('average_costs', ascending=False) #!!!
```
